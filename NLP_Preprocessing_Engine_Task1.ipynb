# Task 1: Conceptual Understanding

## 1. Difference between "Love" and "love" in NLP

In NLP, "Love" and "love" are treated as different tokens unless lowercasing is applied. Lowercasing helps treat them as the same word and reduces duplication in vocabulary.

## 2. What happens if stopwords are not removed?

If stopwords are not removed:
- Dataset size increases
- Model training becomes slower
- Noise increases
- Important keywords become harder to detect

Example stopwords: is, the, am, are, was

## 3. Two scenarios where removing stopwords is harmful

Example 1: Sentiment Analysis  
Sentence: "I am not happy"  
Removing "not" changes the meaning completely.

Example 2: Question Answering Systems  
Sentence: "What is the time?"  
Removing "what" breaks sentence meaning.

## 4. Difference between stemming and lemmatization

Stemming cuts word endings but may produce incorrect words.  
Example: playing → play

Lemmatization converts words into meaningful root words using a dictionary.  
Example: playing → play

```python
import re

def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove email-like patterns
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove special characters / emojis / punctuation
    text = re.sub(r"[^a-z\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenize
    tokens = text.split()

    # Remove short tokens, but keep meaningful words
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
all_tokens = []

for sentence in test_sentences:

    tokens, clean_sentence = preprocess_text(sentence)

    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    avg_length = sum(len(word) for word in tokens)/total_tokens if total_tokens>0 else 0

    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length,2))

    print("-"*50)

    all_tokens.extend(tokens)
```

```python
from collections import Counter

counter = Counter(all_tokens)

print("Top 10 most frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

```python
from collections import Counter

counter = Counter(all_tokens)

print("Top 10 most frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:

        tokens, clean_sentence = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)
print(result)
```

```python
print(preprocess_text(""))
print(preprocess_text("😍😍😍"))
print(preprocess_text("123456"))
```
