# Sentiment Analysis using NLP Pipeline & Machine Learning Models

# NLP Preprocessing Pipeline Implementation

```python
import pandas as pd

url = "https://raw.githubusercontent.com/dD2405/Twitter_Sentiment_Analysis/master/train.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.shape
```

```python
df['label'].value_counts()
```

```python
df.head(5)
```

```python
import re
import nltk

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):

    # lowercase
    text = text.lower()

    # remove urls
    text = re.sub(r"http\S+|www\S+", "", text)

    # remove punctuation & special characters
    text = re.sub(r"[^a-z\s]", "", text)

    # tokenize
    words = text.split()

    # remove stopwords
    words = [word for word in words if word not in stop_words]

    # lemmatization
    words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
df['clean_text'] = df['tweet'].apply(preprocess_text)

df[['tweet','clean_text']].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bow = CountVectorizer(max_features=5000)

X_bow = bow.fit_transform(df['clean_text'])

X_bow.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000)

X_tfidf = tfidf.fit_transform(df['clean_text'])

X_tfidf.shape
```

```python
y = df['label']
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train_bow, X_test_bow, y_train, y_test = train_test_split(
    X_bow, y, test_size=0.2, random_state=42
)

X_train_tfidf, X_test_tfidf, y_train_tfidf, y_test_tfidf = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train_tfidf, y_train_tfidf)

lr_pred = lr_model.predict(X_test_tfidf)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Logistic Regression Results")
print("Accuracy:", accuracy_score(y_test_tfidf, lr_pred))
print("Precision:", precision_score(y_test_tfidf, lr_pred, average='weighted'))
print("Recall:", recall_score(y_test_tfidf, lr_pred, average='weighted'))
print("F1 Score:", f1_score(y_test_tfidf, lr_pred, average='weighted'))
```

```python
from sklearn.naive_bayes import MultinomialNB

nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train_tfidf)

nb_pred = nb_model.predict(X_test_tfidf)
```

```python
print("Naive Bayes Results")
print("Accuracy:", accuracy_score(y_test_tfidf, nb_pred))
print("Precision:", precision_score(y_test_tfidf, nb_pred, average='weighted'))
print("Recall:", recall_score(y_test_tfidf, nb_pred, average='weighted'))
print("F1 Score:", f1_score(y_test_tfidf, nb_pred, average='weighted'))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_tfidf, y_train_tfidf)

dt_pred = dt_model.predict(X_test_tfidf)
```

```python
print("Decision Tree Results")
print("Accuracy:", accuracy_score(y_test_tfidf, dt_pred))
print("Precision:", precision_score(y_test_tfidf, dt_pred, average='weighted'))
print("Recall:", recall_score(y_test_tfidf, dt_pred, average='weighted'))
print("F1 Score:", f1_score(y_test_tfidf, dt_pred, average='weighted'))
```

```python
lr_model_bow = LogisticRegression(max_iter=1000)

lr_model_bow.fit(X_train_bow, y_train)

lr_pred_bow = lr_model_bow.predict(X_test_bow)
```

```python
print("Logistic Regression with BoW Results")

print("Accuracy:", accuracy_score(y_test, lr_pred_bow))
print("Precision:", precision_score(y_test, lr_pred_bow, average='weighted'))
print("Recall:", recall_score(y_test, lr_pred_bow, average='weighted'))
print("F1 Score:", f1_score(y_test, lr_pred_bow, average='weighted'))
```

```python
results = {
    "Model": [
        "Logistic Regression (TF-IDF)",
        "Naive Bayes (TF-IDF)",
        "Decision Tree (TF-IDF)",
        "Logistic Regression (BoW)"
    ],
    "Accuracy": [
        accuracy_score(y_test_tfidf, lr_pred),
        accuracy_score(y_test_tfidf, nb_pred),
        accuracy_score(y_test_tfidf, dt_pred),
        accuracy_score(y_test, lr_pred_bow)
    ],
    "Precision": [
        precision_score(y_test_tfidf, lr_pred, average='weighted'),
        precision_score(y_test_tfidf, nb_pred, average='weighted'),
        precision_score(y_test_tfidf, dt_pred, average='weighted'),
        precision_score(y_test, lr_pred_bow, average='weighted')
    ],
    "Recall": [
        recall_score(y_test_tfidf, lr_pred, average='weighted'),
        recall_score(y_test_tfidf, nb_pred, average='weighted'),
        recall_score(y_test_tfidf, dt_pred, average='weighted'),
        recall_score(y_test, lr_pred_bow, average='weighted')
    ],
    "F1 Score": [
        f1_score(y_test_tfidf, lr_pred, average='weighted'),
        f1_score(y_test_tfidf, nb_pred, average='weighted'),
        f1_score(y_test_tfidf, dt_pred, average='weighted'),
        f1_score(y_test, lr_pred_bow, average='weighted')
    ]
}

comparison_df = pd.DataFrame(results)
comparison_df
```

# Model Comparison & Insights

## Best Preprocessing Steps

Lowercasing, removing punctuation, removing stopwords, tokenization, and lemmatization helped clean the dataset and improve model performance by reducing noise and preserving meaningful words.

## Best Vectorization Technique

Both Bag of Words and TF-IDF were applied. TF-IDF performed better than Bag of Words because it gives higher importance to meaningful words and lowers the weight of very common words.

## Best Performing Model

Logistic Regression with TF-IDF performed the best among the tested models based on Accuracy and F1-score. Logistic Regression with Bag of Words also performed well, but TF-IDF gave slightly better representation of text.

## Trade-offs Observed

Naive Bayes trained very fast and worked well for text data, but its performance was slightly lower than Logistic Regression.
Decision Tree was easy to understand but less effective for high-dimensional text features.
Bag of Words was simple and effective, while TF-IDF gave better feature quality for sentiment classification.
